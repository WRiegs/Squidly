# De-duplicate the datasets 

Datasets downloaded from  https://github.com/samgoldman97/enzyme-datasets

```python
import pandas as pd
import os
files = os.listdir('../enzyme-datasets/data/processed')
for f in files:
    try:
        df = pd.read_csv(f'../enzyme-datasets/data/processed/{f}', index_col=0)
        value_column = df.columns[-1]
        # Sort by the value and keep the highest one
        df = df.sort_values(by=value_column, ascending=False)
        df = df.drop_duplicates(subset='SEQ', keep='first')
        df.to_csv(f'data/{f}')
        print(f, len(f))
    except:
        print(f)
df
```

```python
# Now we want to also have the worst one
import pandas as pd
import os
files = os.listdir('../enzyme-datasets/data/processed')
for f in files:
    try:
        df = pd.read_csv(f'../enzyme-datasets/data/processed/{f}', index_col=0)
        value_column = df.columns[-1]
        # Sort by the value and keep the highest one
        df = df.sort_values(by=value_column, ascending=True)
        df = df.drop_duplicates(subset='SEQ', keep='first')
        df.to_csv(f'data/negative_{f}')
        print(f, len(f))
    except:
        print(f)
df
```
